* The number of inputs of a single neuron will depend on the number of the previous layer's neurons  
* The biases will be one for each neourn on the layer 


```python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        # Initialize weights and biases
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons) #dot product of inputs X weights (the shape is specified so I don't have to transpose the weights)
        self.biases = np.zeros((1, n_neurons))
    
    def forward(self, inputs):
        # Calculate output values from inputs, weights and biases (Prediction)
        self.output = np.dot(inputs, self.weights) + self.biases
```

```python
#Dataset
X, y = spiral_data(samples=100, classes=3)
dense1= Layer_Dense(2,3)
dense1.forward(X)
print(dense1.output[:5]) #print the first 5 
```
