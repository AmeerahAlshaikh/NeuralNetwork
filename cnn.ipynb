```python
import numpy as np
```

```python
inputs = [[0,2,4,1], [0.3,-1,-0.6,3], [0.2,-2,0,1]] #3 batches of data
weights1 = [[0.2,0.8,-0.5,1], [0.5,-0.91,0.26,-0.5], [-0.26,-0.27,0.17,0.87], [2,-3,1,0.8]] #4 neurons
np.shape(np.array(weights1).T)
```

```python
np.shape(inputs)
```

```python
biases1 = [2,3,0.5,1] 
weights2 = [[0.1,-0.14,0.5,1], [-0.5,0.12,-0.33,2], [-0.44,0.73,-0.13,-3]] #3 neurons expecting (4) inputs
biases2 = [-1,2,-0.5] 

layer1_outputs = np.dot(inputs,np.array(weights1).T) + biases1
layer2_outputs = np.dot(layer1_outputs,np.array(weights2).T) + biases2 #the output of the pervious layer will be the input of the next layer
```

```python
weights3= [[0.3,0.2,-0.5], [0.1,-0.4,0.6]] #2 neurons expecting (3) inputs
biases3= [2,-1] 

layer3_outputs = np.dot(layer2_outputs,np.array(weights3).T) + biases3
```

```python
weights4 = [[0.2,0.8], [-0.5,1.0]] #2 neurons expecting (2) inputs
biases4 = [1,-1]    
layer4_outputs = np.dot(layer3_outputs,np.array(weights4).T) + biases4
```

* The number of inputs of a single neuron will depend on the output of the previous one
* The biases will be one for each neourn on the layer
